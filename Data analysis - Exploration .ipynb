```python
# Fusionner les colonnes "Name" et "Description" dans une colonne "Combined"
df["DescriptionN"] = df["Name"] + " " + df["Description"]

# Afficher le dataframe avec la nouvelle colonne
print(df["DescriptionN"])
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words('english'))

# Supprimer les mots vides de la colonne 'DescriptionN' du DataFrame
df['DescriptionN'] = df['DescriptionN'].apply(lambda x: ' '.join([word for word in word_tokenize(x) if word.lower() not in stop_words]))

# Compter les mots les plus courants
all_descriptions = ' '.join(df['DescriptionN'])
tokens = word_tokenize(all_descriptions)
word_counts = nltk.FreqDist(tokens)
most_common_words = word_counts.most_common()

# Filtrer les mots vides des mots les plus courants
filtered_common_words = [(word, count) for word, count in most_common_words if word.lower() not in stop_words]

# Afficher les mots les plus courants filtrés
for word, count in filtered_common_words:
    print(word, count)
```

# Courbe d'évolution des pays utilisant les projets AI ou ML

```python
import matplotlib.pyplot as plt

# Calculer le nombre d'occurrences de chaque pays
country_counts = dff['Geographic coverage\nCountry'].value_counts()

# Afficher le nombre de projets par pays
for country, count in country_counts.items():
    print(country + ": " + str(count))

# Sélectionner les pays avec le plus grand nombre de projets
top_countries = country_counts.nlargest(10).index
print(top_countries)

# Filtrer les données pour les pays du top 10
filtered_data = dff[dff['Geographic coverage\nCountry'].isin(top_countries)]

# Group the data by start year and country, and count the occurrences
grouped_data = filtered_data.groupby(['Start Year', 'Geographic coverage\nCountry']).size().unstack()

# Create the stacked area chart
fig, ax = plt.subplots(figsize=(20, 10))

# Plot the stacked area chart
grouped_data.plot.area(ax=ax, stacked=True)

# Add a vertical line at the year 2020
ax.axvline(x=2020, color='black', linestyle='--', label='2020', linewidth=5)

# Set labels and title
ax.set_xlabel("Start Year", fontsize=17)
ax.set_ylabel("Count", fontsize=17)
ax.set_title("Courbe d'évolution des pays utilisant les projets AI ou ML", fontsize=17)

# Set the font size of the legend
ax.legend(loc='upper left', fontsize=17)

# Set the font size of the tick labels on the x-axis
ax.tick_params(axis='x', labelsize=17)
ax.tick_params(axis='y', labelsize=17)

# Show the chart
plt.show()
```

# pourcentage de chaque service de chaque pays

```python
# Créer une fonction pour extraire les services correspondants à chaque pays
def get_services(row):
    services = []
    if row['Improved Public Service']:
        services.append('Improved Public Service')
    if row['Improved Administrative Efficiency']:
        services.append('Improved Administrative Efficiency')
    if row['Open government capabilities']:
        services.append('Open government capabilities')
    return services

# Appliquer la fonction aux données filtrées pour créer une nouvelle colonne 'Services'
dff['Services'] = dff.apply(get_services, axis=1)

# Compter le nombre total de projets par pays
total_projects_by_country = dff['Geographic coverage\nCountry'].value_counts()

# Calculer le pourcentage d'apparition de chaque service par pays
for country in total_projects_by_country.index:
    country_data = dff[dff['Geographic coverage\nCountry'] == country]
    total_projects = total_projects_by_country[country]
    services_percentage = {}
    for services in country_data['Services']:
        for service in services:
            if service in services_percentage:
                services_percentage[service] += 1
            else:
                services_percentage[service] = 1
    for service, count in services_percentage.items():
        percentage = count / total_projects * 100
        print(country, '-', service, ':', percentage)
```
